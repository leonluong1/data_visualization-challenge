```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
df = pd.merge(study_results, mouse_metadata, on=["Mouse ID"], how="left")

# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
len(df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_df = df[df[["Mouse ID", "Timepoint"]].duplicated()]
duplicates = duplicate_df["Mouse ID"].unique()
duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
#duplicates = np.append(duplicates, ["b128"])
def check_duplicates(column, duplicates):
    result = [False] * len(column)
    for duplicate in duplicates:
        result = np.logical_or(result, column == duplicate)
    return result
            
duplicate_check = check_duplicates(df["Mouse ID"], duplicates)
full_duplicate_df = df[duplicate_check]
full_duplicate_df 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = df.drop(full_duplicate_df.index)
clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
idx = pd.IndexSlice
regimen_df = clean_df[["Drug Regimen", "Tumor Volume (mm3)"]].groupby("Drug Regimen")
describe_df = regimen_df.describe().loc[:, idx["Tumor Volume (mm3)", ["mean", "std", "50%"]]]
regimen_vars = regimen_df.var().squeeze()
regimen_sems = regimen_df.sem().squeeze()


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_df = pd.DataFrame({
        "Mean Tumor Volume": describe_df["Tumor Volume (mm3)", "mean"],
        "Median Tumor Volume": describe_df["Tumor Volume (mm3)", "50%"],
        "Tumor Volume Variance": regimen_vars,
        "Tumor Volume Std. Dev.": describe_df["Tumor Volume (mm3)", "std"],
        "Tumor Volume Std. Err.": regimen_sems
})

summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_df2 = regimen_df.agg(["mean", "median", "var", "std", "sem"])
summary_df2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = clean_df.loc[:, ["Mouse ID", "Timepoint"]].groupby("Mouse ID").max()
max_timepoints.reset_index(inplace=True)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
timepoint_df = pd.merge(max_timepoints, clean_df, on=["Mouse ID", "Timepoint"], how="left")
timepoint_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
# I used a dict(str, list) instead of a list so I can store the tumor volumes easier.
tumor_data = {}

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for regimen in regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    tumor_data[regimen] = timepoint_df.loc[timepoint_df["Drug Regimen"] == regimen, "Tumor Volume (mm3)"]


    # Determine outliers using upper and lower bounds
    quartiles = tumor_data[regimen].quantile([.25, .5, .75])
    upperq = quartiles[0.75]
    lowerq = quartiles[0.25]
    iqr = upperq - lowerq
    upper_bound = upperq + iqr*1.5,
    lower_bound = lowerq - iqr*1.5
    print(f"{regimen}'s potential outliers: {tumor_data[regimen].loc[(tumor_data[regimen] > upper_bound) | (tumor_data[regimen] < lower_bound)]}")
    
```

```python
##### Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig, ax = plt.subplots()

data = [tumor_data[regimens[0]], tumor_data[regimens[1]], tumor_data[regimens[2]], tumor_data[regimens[3]]]

ax.boxplot(data, labels=regimens, flierprops = dict(marker = "o", markerfacecolor = "red", markersize=12))
plt.ylim(20, 75)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_id = "l509"
mouse_data = clean_df.loc[clean_df["Mouse ID"] == mouse_id].sort_values(by="Timepoint")
plt.plot(mouse_data["Timepoint"], mouse_data["Tumor Volume (mm3)"])
plt.title(f"Capomulin treatment of mouse {mouse_id}")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
chosen_regimen = "Capomulin"
volume_weight = clean_df.loc[clean_df["Drug Regimen"] == chosen_regimen, ["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]]
mean_volume_weight = volume_weight.groupby("Mouse ID").mean()
x_values = mean_volume_weight["Weight (g)"]
y_values = mean_volume_weight["Tumor Volume (mm3)"]
plt.scatter(x_values, y_values)
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
plt.scatter(x_values, y_values)
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
plt.scatter(x_values, y_values, c="C0")
plt.plot(x_values, regress_values,"r-")

print(f"The correlation between mouse weight and the average tumor volume is {round(rvalue, 2)}")
plt.show()
```
